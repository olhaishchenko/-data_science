```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# У цьому завданні вам потрібно завантажити ось цей датасет. Тут ви знайдете 2 файли - з двовимірним датасетом та датасетом mnist. Для кожного з них застосуйте алгоритм K-means для кластеризації. Щоб знайти оптимальну кількість кластерів, скористайтесь ліктевим методом.
```

```python
df1 = pd.read_csv("/Users/olha/PycharmProjects/pythonProject1/data 2/data_2d.csv", delimiter=",", header=None)
```

```python
X = df1[[1,2]]
```

```python
X_norm = StandardScaler().fit_transform(X)
```

```python
cost = []

for k in range(1, 6):
    kmeans = KMeans(n_clusters=k).fit(X_norm)
    cost.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 6), cost, marker=".")
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X[1], X[2])
```

```python
kmeans = KMeans(n_clusters=2).fit(X_norm)
plt.figure(figsize=(8, 8))
plt.scatter(X_norm[:,0], X_norm[:,1], c=kmeans.labels_)
```

```python
kmeans1 = KMeans(n_clusters=3).fit(X_norm)
plt.figure(figsize=(8, 8))
plt.scatter(X_norm[:,0], X_norm[:,1], c=kmeans1.labels_)
```
