Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import pandas as pd
import numpy as np

df = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C",
                   match="Коефіцієнт народжуваності в регіонах України")[0]
```

Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', np.nan, inplace=True)
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df['2014'] = df['2014'].astype(float, copy=True)
df['2019'] = df['2019'].astype(float, copy=True)
```

```python
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().mean()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df_new = df.drop([27], axis=0)
df_new
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df_new = df_new.fillna({"1950": df_new["1950"].mean(), "1960": df_new["1960"].mean(), "1970": df_new["1970"].mean(), "2014": df_new["2014"].mean(), "2019": df_new["2019"].mean()})
df_new
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
average_2019 = df_new["2019"].mean()
df_new[df_new["2019"] > average_2019]["Регіон"].tolist()
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df_new.loc[df_new["2014"].idxmax()]["Регіон"]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt

plt.barh(df_new["Регіон"], df_new["2019"])
```

```python
plt.pie(df_new["1950"], labels=df_new["Регіон"], autopct="%.2f%%", pctdistance=1.15, labeldistance=1.55);
```

```python
df_krym = df_new[df_new["Регіон"]=='Крим']
df_dnipro = df_new[df_new["Регіон"]=='Дніпропетровська']
df_chern = df_new[df_new["Регіон"]=='Чернігівська']
```

```python
x=[1950, 1960 , 1970, 1990, 2000, 2012, 2014, 2019]
y_krym = [df_krym['1950'], df_krym['1960'], df_krym['1970'], df_krym['1990'], df_krym['2000'], df_krym['2012'], df_krym['2014'], df_krym['2019']]
y_dnipro = [df_dnipro['1950'], df_dnipro['1960'], df_dnipro['1970'], df_dnipro['1990'], df_dnipro['2000'], df_dnipro['2012'], df_dnipro['2014'], df_dnipro['2019']]
y_chern =[df_chern['1950'], df_chern['1960'], df_chern['1970'], df_chern['1990'], df_chern['2000'], df_chern['2012'], df_chern['2014'], df_chern['2019']]
fig, axs = plt.subplots()
axs.plot(x, y_krym , label='Krym')
axs.plot(x, y_dnipro , label='Dnipro')
axs.plot(x, y_chern , label='Chernihiv')
plt.xlabel('Year', fontsize='small', color='midnightblue')
plt.ylabel('Birth', fontsize='small', color='midnightblue')
plt.legend();
```
